# 분석 제안서

## 1조 : 조영은, 김동현, 김태섭

1. 개요
- 공모전 : 2024 문화 디지털혁신 및 문화데이터 활용 공모전
- 참가분야 : 데이터분석

분석의 목적 :
 지역축제 불균형 원인 분석과 축제간 격차 해소를 위한 방안

문제 설명 :
 지역축제의 불균등한 배분 문제는 경제적 격차, 문화적 동질화, 사회적 불평등으로 이어지며, 활성화 된 지역의 성공적인 축제가 덜 활성화 된 지역의 축제를 압도하여 이러한 문제를 더욱 악화시킨다. 이러한 문제를 해결하기 위해서는 재정 지원, 인프라 개발, 인재 육성, 효과적인 마케팅 및 커뮤니티 참여 등의 솔루션이 필요하다.
- 문화데이터 활용 및 데이터 목적에 부합한 해석 & 해결방안 서술
- 분석 외 새로운 지표(데이터 시각화) 혹은 분석 내용에 대한 의견 제안
- 문화데이터 분석을 통한 정책 현안 해결 또는 분석 인사이트 도출


2. 데이터 수집
- 데이터 출처:
  - 공공데이터포털, 한국관광데이터랩, 국토교통부 레일포털
- 수집 방법:
  - 공공데이터수집, 한국관광데이터랩, 국토교통부 레일포털
- 수집된 데이터의 구조 및 특성 설명:
    - 지역 축제 데이터, 서울, 대구 지하철 데이터, 경상북도 경주시, 안동시, 포항 남구 방문자 수 데이터 셋.
    - 모두 .csv 확장자
    - 지역 축제 데이터의 경우 : 1719 행 과 18개 열로 이루어져있으며, 지역별 축제 기간 및 개최지, 전화번호 및 주소가 작성.
    - 서울, 대구 지하철 데이터의 경우 : 호선 별로, 위도 및 경도가 작성, 대구 지하철 데이터는 도로명 주소만 작성.
    - 경상북도 경주시, 안동시, 포함 남구 방문자 수 데이터의 경우 : 22 ~ 23년도 방문자 수가 작성.

3. 데이터 전처리 (필요 시 작성. 데이터가 확보된 경우 작성. 아래의 것을 모두 작성할 필요는 없으며, 데이터에 맞춰서 작성함)

결측치 처리 : 
- 지역 축제 데이터의 필요없는 NaN 값이 포함이 되어 있는 데이터기준일자, 제공기관코드, 제공기관명, 홈페이지 주소, 전화번호, 축제 내용의 열을 제거 예정
  위도와 경도 데이터의 NaN 값이 포함 되어있는 행을 제거 예정

- 지하철 위치 데이터의 값들은 누락된 데이터가 없어서 처리하지 않음

- 경상북도 경주시, 안동시, 포항 남구 방문자 수 데이터의 값들은 누락된 결측치가 없음

이상치 처리 :
- 이상치 발견하지 못함.

데이터 정규화/표준화 :
- 지도 상에 표시하는 것이기에 표준화 진행할 필요성 없음

피처 엔지니어링 (변수 선택 및 변환) : 
- 경상북도 경주시, 안동시, 포항 남구 방문자 수 데이터 셋을 결합할 예정

4. 데이터 탐색 및 시각화 (데이터가 확보된 경우 작성. 아래의 것을 모두 작성할 필요는 없으며, 데이터에 맞춰서 작성함)

기술통계 분석 : 상관성이 보이는 데이터가 따로 없음

데이터 분포 확인 : isna(), sum(), 함수로 NaN 값이 들어있는 행 확인

상관관계 분석 : 어느 지역 축제가 가장 많이 열리는 지 지역 비교, 지도 및 그래프로 상관 관계 분석

중요 변수 시각화 : 지역 축제 좌표를 지도에 표시 예정, 히트맵으로 어느 지역이 많고 적은지 확인할 예정
