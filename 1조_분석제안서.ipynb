# 분석 제안서

## 1조 : 조영은, 김동현, 김태섭

1. 개요
- 공모전 : 교육 공공데이터 분석활용대회
- 참가분야 : 데이터분석

분석의 목적 :
 지역별 교육 기회의 현황 파악, 불균형 요인 분석, 개선 방안 모색

문제 설명 :
 지역별로 교육 기회의 분포가 고르지 않다는 문제는 사회적 불평등을 심화시킬 수 있으며, 교육 기회의 불균형은 학생들의 학업 성취도, 진로 선택의 제한, 사회적 이동성 저하 등 여러 부정적인 결과를 초래할 수 있다. 따라서, 지역별 교육 기회의 불균형을 정확히 파악하고 이에 대한 해결책 및 방안을 모색.
- 교육 공공데이터 활용 및 데이터 목적에 부합한 해석 & 해결방안 서술
- 분석 외 새로운 지표(데이터 시각화) 혹은 분석 내용에 대한 의견 제안
- 분석한 결과 및 새로운 아이디어 제시

2. 데이터 수집
- 데이터 출처: 공공데이터포털 활용
- 수집 방법: 공공데이터수집
- 수집된 데이터의 구조 및 특성 설명: 데이터셋을 수집하고 있는 중이기 때문에, 정형 데이터, 비정형및 반정형 데이터 셋을 구할 예정, 확장자 파일은 CSV, Excel, json 등으로 구할 예정.
